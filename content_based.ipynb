```
import pandas as pd
import numpy as np
```

```
##Step 1: Read CSV File
df = pd.read_csv("movie_dataset.csv")
df.head()
```

```
# combine features
def combine_features(row):
  try:
    ## Handling NaN values i.e replacing with empty string
    row.fillna('', inplace=True)
    return row['keywords'] + " " + row['genres'] + " " + row['cast'] + " " + row['director']
  except:
    print(row)

df['combined_features'] = df.apply(combine_features,axis=1)
```

```
df.iloc[0]['combined_features']
```

```
## find the count matrix and cosin similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

```
##Step 5: Compute the Cosine Similarity based on the count_matrix
cosine_sim = cosine_similarity(count_matrix) 
```

```
sim_df = pd.DataFrame(cosine_sim,index=df.title,columns=df.title)
```

```
sim_df.head()
```
